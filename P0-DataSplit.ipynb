```python
white = pd.read_csv("white.tsv", sep="\s+")[["FID"]].rename(columns={"FID": "eid"})
white
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(white, test_size=0.5, random_state=0)
train, test = train.reset_index(drop=True), test.reset_index(drop=True)
train.to_csv("white_train.csv", index=False)
test.to_csv("white_test.csv", index=False)
```

##  划分男女

```python
train = pd.read_csv("white_train.csv")
```

```python
covs = pd.read_csv("/home/xutingfeng/ukb/ukbData/phenotypes/regenie.cov", sep="\s+")
covs["eid"] = covs["FID"]
covs
```

```python
train["eid"] = train["eid"].astype(str)
covs["eid"] = covs["eid"].astype(str)
```

```python
female = covs.query("inferred_sex == 0 and eid in @train.eid")[
    ["FID", "IID"]
].reset_index(drop=True)
male = covs.query("inferred_sex == 1 and eid in @train.eid")[
    ["FID", "IID"]
].reset_index(drop=True)
# female
print(female.shape, male.shape)
```

```python
male.to_csv("white_train_male.tsv", sep="\t", index=False, header=False, na_rep="NA")
female.to_csv(
    "white_train_female.tsv", sep="\t", index=False, header=False, na_rep="NA"
)
```
